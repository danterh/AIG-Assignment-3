```
import numpy as np
from matplotlib import pyplot as plt
import math
```

```
r = 0
numerical_factor = 5
drag_coefficient = 4
lift_coefficient = 3
velocity = 20
unit_vector = np.array([[500, 900, 300],
                        [456, 50, 200],
                        [797, 876, 987]])
velocity_vector = np.array([[309, 1, 190],
                           [256, 512, 223],
                           [964, 872, 723]])
gravity = 9.8
```

```
print(unit_vector)
```

```
r = ((numerical_factor*(-1))*drag_coefficient*(velocity**2)*velocity_vector) + (numerical_factor * lift_coefficient * velocity**2 * (unit_vector * velocity_vector)) + gravity
```

```
print(r)
```

```
from mpl_toolkits import mplot3d
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(r[0],r[1],r[2])
# ax.scatter(r[0],r[1],r[2], zdir='z', c= 'red')
plt.show()
```
